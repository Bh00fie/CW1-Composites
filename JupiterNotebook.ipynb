```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import make_interp_spline

x = [0, 1, 2, 3, 4, 5]
y1 = [1, 0.999357, 0.995498, 0.972347, 0.833441, 0]
y2 = [0, 0.2, 0.4, 0.6, 0.8, 1]

x_smooth = np.linspace(min(x), max(x), 300)
spline1 = make_interp_spline(x, y1)
spline2 = make_interp_spline(x, y2)
y1_smooth = spline1(x_smooth)
y2_smooth = spline2(x_smooth)

plt.figure(figsize=(8, 4))
plt.plot(x_smooth, y1_smooth, label='Line 1')
plt.plot(x_smooth, y2_smooth, label='Line 2')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Your Graph Title')
plt.legend()
plt.show()
```
